```python
import json
import os

# Mặc định nếu không tìm thấy file
params = {
    "dataset": "seq-cifar100",
    "lr": 0.03,
    "batch_size": 32,
    "n_epochs": 5
}

if os.path.exists('./params.json'):
    print("Có file params.json, ghi đè tham số...")
    with open('./params.json', 'r') as f:
        params.update(json.load(f))

print(f"Chạy với tham số: {params}")
```

```python
os.chdir('/kaggle/working/NeurIPS2025-LEAR')
cmd = f"""python main_domain.py --dataset {params['dataset']} \
    --list_datasets {params['list_datasets']} \
    --ncls_per_task {params['ncls_per_task']} \
    --model {params['model']} \
    --lr {params['lr']} \
    --batch_size {params['batch_size']} \
    --n_epochs {params['n_epochs']} \
    --num_workers 8 \
    --backbone {params['backbone']} \
    --transform_type {params['transform_type']} \
    --use_bilora {params['use_bilora']} \
    --skip_task_0 {params['skip_task_0']} \
    --bilora_mode {params['bilora_mode']} \
    --apply_bilora_for {params['apply_bilora_for']} """

os.system(cmd)
```
